```python
def calculate_metrics(df):
    metrics = df.group_by(
        pl.col('date').cast(pl.Date),
        pl.col('date').cast(pl.Date).dt.year().alias('year'),
        pl.col('date').cast(pl.Date).dt.month().alias('month'),
        pl.col('date').cast(pl.Date).dt.day().alias('day'),
        pl.col('model')
        ).agg(pl.sum("failure").alias('failure_count'))
    return metrics
```

```python
lazy_df = pl.scan_csv('../Data/source/2013/*.csv')
```

```python
df = calculate_metrics(lazy_df)
```

```python
df.collect().sort(['failure_count', 'year', 'month', 'day', 'model'], descending=True)
```

```python
def write_parquets(df, path: str) -> None:
    df.sink_parquet(path, compression='snappy')
```

```python
metrics_path = '../Data/target/2013/metrics_2013.parquets'
```

```python
write_parquets(df, metrics_path)
```
